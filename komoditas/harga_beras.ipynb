```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("beras_complete.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[['lag_1','lag_7','lag_30']] = df[['lag_1','lag_7','lag_30']].bfill()
df['daily_delta'] = df['daily_delta'].fillna(0)
df['nilai_diff'] = pd.to_numeric(df['nilai_diff'], errors='coerce').fillna(0)
```

```python
df = df.sort_values('Tanggal')
df['rolling_mean_7_custom'] = df['prices'].rolling(window=7, min_periods=1).mean()
df['rolling_mean_30_custom'] = df['prices'].rolling(window=30, min_periods=1).mean()
```
